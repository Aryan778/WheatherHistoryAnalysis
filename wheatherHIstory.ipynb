```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Aryan778/wheatherHistory/251189b94af8c3bc6451163a597228ce2c8a7436/weatherHistory.csv'
weatherHistory = pd.read_csv(url)
weatherHistory.coloumns = ['Formatted Date,Summary,Precip Type,Temperature (C),Apparent Temperature (C),Humidity,Wind Speed (km/h),Wind Bearing (degrees),Visibility (km),Loud Cover,Pressure (millibars),Daily Summary']
weatherHistory.head()
```

```python
weatherHistory.isnull().sum()
```

```python
weatherHistory.info()
```

```python
weatherHistory['Formatted Date']=pd.to_datetime(weatherHistory['Formatted Date'],utc=True)
```

```python
weatherHistory['Formatted Date']
```

```python
weatherHistory=weatherHistory.set_index('Formatted Date')
weatherHistory.head()
```

```python
data_columns = ['Apparent Temperature (C)','Humidity']
df_monthly_mean = weatherHistory[data_columns].resample("MS").mean()
df_monthly_mean.head()
```

```python
df_monthly_mean
```

```python
df1=df_monthly_mean[df_monthly_mean.index.month==4]
print(df1)
df1.dtypes
```

```python
import matplotlib.dates as mdates
from datetime import datetime
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(df1.loc['2006-04-04':'2016-04-01','Apparent Temperature (C)'], marker='o',linestyle='-',label='Apparent Temperature (C)')
ax.plot(df1.loc['2006-04-04':'2016-04-01','Humidity'], marker='o', linestyle='-',label='Humidity')
ax.xaxis.set_major_formatter(mdates.Dateformatter('%v-%m'))
ax.legend(loc = 'center right')
ax.set_xlabel('Month of April')
```
